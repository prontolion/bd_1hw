### I. Таблица Транзакций

1) Находится в 1НФ, поскольку выполняютя все условия:
    - Повторных строк в таблице нет (отличительная черта - ID транзакции);
    - Все атрибуты простые;
    - Все домены содержат скалярные значения.

2) Не может быть преобразована в 3НФ по следующей причине:
    - Ключевые атрибуты (Цены/Затраты) зависят транзитивно от одного лишь первичного ключа, иными словами, присутствует неприводимость для набора параметров, которые вместе формируют цены и затраты.

В данном случае первичным ключём может выступать ID продукта, и достаточно выбрать его равный 7, чтобы увидеть транзитивность и неприводимость.
Соответственно, пробуем привести к 2НФ.

Устанавливаем две таблицы со следующими логическими связями:

- ID транзакции | ID продукта, Бренд, Линейка продукта, Класс продукта, Размер продукта ---> Цена, Стандартные расходы;
- ID транзакции | ID клиента, Дата транзакции, Доставка онлайн, Статус доставки (всё, что не влияет на цену/расходы).

Ключом для них выступает столбец ID транзакции.

### II. Таблица Клиентов

1) Находится в 1НФ, поскольку выполняютя все условия:
    - Повторных строк в таблице нет (отличительная черта - ID клиента);
    - Все атрибуты простые;
    - Все домены содержат скалярные значения.

2) Эту таблицу можно привести к 2НФ, а позже к 3НФ, так как все атрибуты зависят от ID клиента, однако так как атрибутов слишком много, чтоб не создавать большое количество БД, имеет смысл сгруппировать их по основным критериям, описанным ниже.

Устанавливаем три таблицы со следующими логическими связями:

- ID клиента | Имя, Фамилия, Пол, Дата Рождения, Индикатор смерти (основная информация по клиенту);
- ID клиента | Название профессии, Название индустрии, Категория достатка, Наличие машины, Оценка собственности (состояние клиента);
- ID клиента | Адрес, Почт. индекс, Штат, Страна (адрес проживания клиента).

Ключом для них выступает столбец ID клиента.
